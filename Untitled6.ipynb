```python
### Q30.

Create a dictionary where:

- Key = Student Name
- Value = Marks

students = {
    "Rahul": 85,
    "Priya": 92,
    "Amit": 78,
    "Neha": 88,
    "Siva": 75
}

print("Students who scored more than 80 marks:")

for name, marks in students.items():
    if marks > 80:
        print(name, ":", marks)
```
